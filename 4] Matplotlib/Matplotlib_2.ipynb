```python
"""
3D Plots
a. Creating 3D Line and Scatter Plots
3D plots are useful for visualizing data with three dimensions. 
We can use the Axes3D module from matplotlib to create these plots.
"""
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data for 3D line
x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)

# Data for 3D scatter
x_scatter = np.random.standard_normal(100)
y_scatter = np.random.standard_normal(100)
z_scatter = np.random.standard_normal(100)

fig = plt.figure()

# Creating 3D line plot
ax = fig.add_subplot(121, projection='3d')
ax.plot(x, y, z, label='3D Line')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.legend()

# Creating 3D scatter plot
ax = fig.add_subplot(122, projection='3d')
ax.scatter(x_scatter, y_scatter, z_scatter, color='r', marker='o', label='3D Scatter')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.legend()

plt.show()



"""
3D Surface Plots
3D Surface plots represent data as a 3D surface, which is especially useful for functions of two variables.
"""
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate data
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Creating the surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()

"""
Customizing 3D Plots
Customization in 3D plots can involve changing the color, adding labels, modifying the view angle, etc.
"""
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for 3D surface
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Surface plot with customization
surface = ax.plot_surface(X, Y, Z, cmap='inferno', edgecolor='none')

# Customizing the view angle
ax.view_init(elev=30, azim=120)
#Sets the elevation and azimuth angles to control the viewing angle.

# Adding a color bar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()

```

```python
"""
Plotting Time Series Data
Plotting time series data involves representing data points in time order.
"""
import pandas as pd
import matplotlib.pyplot as plt

# Creating a sample time series
dates = pd.date_range('20230101', periods=100)
data = pd.Series(np.random.randn(100), index=dates)

# Plotting time series data
data.plot()
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()


"""
Formatting Dates on X-Axis
Formatting dates on the x-axis enhances the readability of time series plots.
"""
fig, ax = plt.subplots()

# Plotting
data.plot(ax=ax)

# Formatting the x-axis dates
ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))
fig.autofmt_xdate()

plt.title('Time Series with Formatted Dates')
plt.show()

"""
Handling Missing Data in Time Series
Time series data may have missing values, which can be handled by interpolation or filling.
"""
# Introduce missing data
data_missing = data.copy()
data_missing[10:20] = np.nan

# Plot with missing data
data_missing.plot(label='Original with NaNs')

# Handling missing data with interpolation
data_filled = data_missing.interpolate()
data_filled.plot(label='Interpolated')

plt.legend()
plt.title('Handling Missing Data in Time Series')
plt.show()

```

```python
"""
Creating Contour and Filled Contour Plots
Contour plots show the level curves of a 3D surface on a 2D plane.
"""
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Creating contour plot
plt.contour(X, Y, Z, levels=10, colors='black')

# Creating filled contour plot
plt.contourf(X, Y, Z, levels=10, cmap='viridis') #levels=10--> Defines the number of contour levels.

plt.colorbar()
plt.title('Contour and Filled Contour Plots')
plt.show()

"""
Customizing Contour Levels
Customizing contour levels helps in highlighting specific ranges.
"""
# Customizing contour levels
levels = np.arange(-1, 1.1, 0.2)

# Creating contour plot
contours = plt.contour(X, Y, Z, levels=levels, colors='black') #levels=np.arange(-1, 1.1, 0.2): Specifies custom contour levels.


# Adding labels to contours
plt.clabel(contours, inline=True, fontsize=8)

# Filled contour plot
plt.contourf(X, Y, Z, levels=levels, cmap='coolwarm')

plt.colorbar()
plt.title('Customized Contour Levels')
plt.show()
```

```python
"""
Plotting Images using imshow()
The imshow() function in matplotlib is used to display images. It interprets the data as an image and displays it accordingly.
"""
import matplotlib.pyplot as plt
import numpy as np

# Creating a sample image (2D array)
image = np.random.random((10, 10))

# Plotting the image
plt.imshow(image, cmap='gray')
plt.title('Image Display with imshow()')
plt.show()

"""
Modifying Image Properties (Colormap, Interpolation)
You can modify various properties of the image, such as the colormap and interpolation, to change how the image is displayed.

a. Changing the Colormap
The colormap controls the colors used to map the pixel values of the image.
"""
# Displaying the image with a different colormap
plt.imshow(image, cmap='viridis')
plt.title('Image with Viridis Colormap')
plt.show()

"""
Changing the Interpolation
Interpolation controls how pixel values are interpolated between the actual data points.
"""
# Displaying the image with different interpolation methods
plt.imshow(image, cmap='viridis', interpolation='nearest')
plt.title('Image with Nearest Interpolation')
plt.show()
```

```python
"""
bilinear': Linear interpolation, which smooths the image.
bicubic': Cubic interpolation, which smooths even further.

Adding Colorbars to Plots
Colorbars provide a reference for the mapping between data values and colors in the image.
"""
# Plotting the image with a colorbar
plt.imshow(image, cmap='viridis', interpolation='bilinear')
plt.colorbar()
plt.title('Image with Colorbar')
plt.show()



import matplotlib.pyplot as plt
import numpy as np

# Creating a sample image
image = np.random.random((10, 10))

# Plotting the image with modified properties and a colorbar
plt.imshow(image, cmap='plasma', interpolation='bicubic')
plt.colorbar()
plt.title('Image with Plasma Colormap and Bicubic Interpolation')
plt.show()

```
